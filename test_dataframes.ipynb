```python
import pandas as pd
import seaborn as sns
from helpers.dataframes import *

df = pd.read_csv('test_datasets/wiki-RfA.csv')
df.dropna(inplace=True)
display(df)
```

```python
# "the number of RfAs per year"
df_ = number_of_unique_x_per_y(df, x_col='TGT', y_col='YEA')
display(df_)
sns.lineplot(data=df_, x='YEA', y='Number of unique TGT per YEA')
```

```python
# "the overall fraction of positive votes per year"
df_ = fraction_of_x_per_y(df, x_col='VOT', y_col='YEA', x_val=1)
display(df_)
sns.lineplot(data=df_, x='YEA', y='Fraction of VOT=1 per YEA')
```

```python
# "the average number of votes per RfA per year"
df_ = average_number_of_x_per_y1_per_y2(df, x_col='VOT', y1_col='TGT', y2_col='YEA')
display(df_)
sns.lineplot(data=df_, x='YEA', y='Average number of VOT per TGT per YEA')
```
